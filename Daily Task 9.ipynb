#### Cricket Program:

```python
import random

def cricket_match(balls, wickets):
    print(f"Welcome to the IPL T20 Cricket Match !!!")
    print("Teams are ready for an exciting clash. Let the match begin!\n")

    total_score = 0
    total_wickets = 0

    while balls > 0 and wickets > 0:
        batting_result = random.choice([1, 2, 3, 4, 6, "OUT"])

        if batting_result in [1, 2, 3, 4, 6]:
            total_score += batting_result
            if batting_result == 6:
                print("SIX! A maximum! The crowd erupts with excitement!")
            elif batting_result == 4:
                print("FOUR! Beautifully timed boundary. The crowd cheers!")
            else:
                print(f"{batting_result} runs scored!")

        else:
            print("OUT! Umpire decision")
            umpire_result = random.choice(["Bowled", "Run Out", "Not Out", "lbw", "Stumped", "Hit wicket", "No Ball"])

            if umpire_result in ["Bowled", "Run Out", "lbw", "Stumped", "Hit wicket"]:
                wickets -= 1
                print(f"Umpire says {umpire_result}! Wicket lost. The crowd goes silent.")
            elif umpire_result == "Not Out":
                print("Umpire says Not Out! The batsman survives the appeal.")
            elif umpire_result == "No Ball":
                total_score += 1
                print("Umpire says No Ball! Free hit and 1 run added to the total.")
                balls += 1  

        balls -= 1
        print(f"Balls left: {balls}, Wickets remaining: {wickets}, Total Score: {total_score}\n")

    target_score = total_score + 1

    if wickets == 0:
        print(f"All wickets lost! Game Over.")
    else:
        print(f"All balls bowled! Game Over.")

    print(f"Final score is {total_score} runs for {total_wickets} wickets.")
    print(f"Target Score is {target_score}")

cricket_match(120, 10)
```
